# Why is $\int \frac{dy}{dx} \, dx = y(x) + C$

Lets take the $y = x^2$ example.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(0, 10, 0.1)
y = x**2

dx = x[1] - x[0] # 0.1 

dydx = np.gradient(y, x)
dydx_int = np.cumsum(dydx) * dx

plt.plot(x, y, 'r', label = 'f(x)')
plt.plot(x, dydx, 'blue', label = 'dy/dx')
plt.plot (x, dydx_int, 'g--', linewidth = 2.5 , label = 'int(dy/dx)')

for i in range(len(x) - 1):
    plt.bar(x[i], dydx[i], width = dx, align = 'edge', alpha = 0.4, edgecolor = 'k', color = 'lightblue')

plt.legend()
```
