# Learning the modern version of plt.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Instead of plt.plot, do:

x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a figure with two rolls and two columns:

fig, ax = plt.subplots(2, 2, figsize = (8, 6))
```

```python
fig, ax = plt.subplots(2, 2, figsize = (8, 6))

ax[0,0].plot(x, y)
ax[0,0].set_title('Top Left')

ax[0,1].plot(x, np.cos(x))
ax[0,1].set_title('Top Right')

ax[1,0].plot(x, np.tan(x))
ax[1,0].set_title('Bottom Left')

ax[1,1].plot(x, np.exp(x))
ax[1,1].set_title('Bottom Right')

plt.tight_layout()
plt.show()
```
